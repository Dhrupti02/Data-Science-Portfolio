```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
ibm = pd.read_csv('IBMStock.csv')
boeing = pd.read_csv('BoeingStock.csv')
ge = pd.read_csv('GEStock.csv')
cock = pd.read_csv('CocaColaStock.csv')
procter = pd.read_csv('ProcterGambleStock.csv')
```

```python
ibm_df = pd.DataFrame(ibm)
boeing_df = pd.DataFrame(boeing)
ge_df = pd.DataFrame(ge)
cock_df = pd.DataFrame(cock)
procter_df = pd.DataFrame(procter)
```

```python
### checking for missing values

ibm_df.isnull().sum()
boeing_df.isnull().sum()
ge_df.isnull().sum()
cock_df.isnull().sum()
procter_df.isnull().sum()
```

```python
#Converting date string into date object

ibm_df['Date'] = pd.to_datetime(ibm_df['Date'], format = "%m/%d/%y")
boeing_df['Date'] = pd.to_datetime(boeing_df['Date'], format = "%m/%d/%y")
ge_df['Date'] = pd.to_datetime(ge_df['Date'], format = "%m/%d/%y")
cock_df['Date'] = pd.to_datetime(cock_df['Date'], format = "%m/%d/%y")
procter_df['Date'] = pd.to_datetime(procter_df['Date'], format = "%m/%d/%y")
```

#### As we've seen, there are no missing values in any dataset. Now we can further do some exploratory data analysis on these datasets. EDA is all about exploring the data by seeking the answers.

```python

sns.lineplot(ibm_df['Date'], ibm_df['StockPrice'], color = 'r')
sns.lineplot(boeing_df['Date'], boeing_df['StockPrice'], color = 'b')
sns.lineplot(ge_df['Date'], ge_df['StockPrice'], color = 'g')
sns.lineplot(cock_df['Date'], cock_df['StockPrice'], color = 'k')
sns.lineplot(procter_df['Date'], procter_df['StockPrice'], color = 'm')
```

Now adding two columns, year and month.

```python
### creating year and month column
ibm_df['Year'] = ibm_df['Date'].dt.year
ibm_df['Month'] = ibm_df['Date'].dt.month_name()
#df['Month'] = df['Month'].apply(lambda x: calender.month_abbr[x])

boeing_df['Year'] = boeing_df['Date'].dt.year
boeing_df['Month'] = boeing_df['Date'].dt.month_name()
#df1['Month'] = df1['Month'].apply(lambda x: calender.month_abbr[x])

ge_df['Year'] = ge_df['Date'].dt.year
ge_df['Month'] = ge_df['Date'].dt.month_name()
#df2['Month'] = df2['Month'].apply(lambda x: calender.month_abbr[x])

cock_df['Year'] = cock_df['Date'].dt.year
cock_df['Month'] = cock_df['Date'].dt.month_name()
#df3['Month'] = df3['Month'].apply(lambda x: calender.month_abbr[x])

procter_df['Year'] = procter_df['Date'].dt.year
procter_df['Month'] = procter_df['Date'].dt.month_name()
```

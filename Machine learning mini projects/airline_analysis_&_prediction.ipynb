## How does the number of people flying between different locations change over time?

```python
## new feature for month/year
new_data['YearMonth'] = new_data['Month'].astype(str)+'/'+new_data['Year'].astype(str)
```

```python
## convert YearMonth into datetime object and store into new feature Month_Year
new_data['Month_Year'] = pd.to_datetime(new_data['YearMonth'])
```

```python
## calculate mean distance
avg_travel = new_data[['Month_Year','Distance']].groupby(['Month_Year'])['Distance'].mean
avg_travel_s = avg_travel()
```

```python
## calculate sum of distance
agg_travel = new_data[['Month_Year','Distance']].groupby(['Month_Year'])['Distance'].sum
agg_travel_s = agg_travel()
```

```python
## calculate count of distance
count_travel = new_data[['Month_Year','Distance']].groupby(['Month_Year'])['Distance'].count
count_travel_s = count_travel()
```

```python
## store these three into new dataframe
new_res = pd.DataFrame({'Average_tr_dist':avg_travel_s,'Total_tr_dist':agg_travel_s,'Flight_counts':count_travel_s})
```

## Can you detect cascading failures as delays in one airport create delays in others?

```python
## get data with departure and arrival delay
dep_d = new_data[new_data.DepDelay > 0]
arr_d = dep_d[dep_d.ArrDelay > 0]
```

```python
## calculate cascading failures and print it
print ('cascading failures in percentage: ', round(arr_d.shape[0]/dep_d.shape[0]*100,5),"%")
```

### There is a 78.21% chance that delay on the departure airport will create a delay on the arrival airport as well, therefore cascading failures do seem to exist.

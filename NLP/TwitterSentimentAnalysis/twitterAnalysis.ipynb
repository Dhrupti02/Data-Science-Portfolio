## Tokenization

```python
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
```

```python
tt = TweetTokenizer()
final_df['Full Text'] = final_df['Full Text'].apply(tt.tokenize)
```

```python
final_df['Full Text']
```

### remove stopwords

```python
stopwords_list = stopwords.words('english')
final_df['tweet_without_stopwords'] = final_df['Full Text'].apply(lambda x: ' '.join([word for word in x if word not in (stopwords_list)]))
```

```python
final_df['tweet_without_stopwords']
```

## Stimming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
```

```python
def tokenize(text):
    return word_tokenize(text)

def stemming(words):
    stem_words = []
    for w in words:
        w = ps.stem(w)
        stem_words.append(w)
    
    return stem_words
```

```python
# apply tokenize function
final_df['Text'] = final_df['tweet_without_stopwords'].apply(tokenize)
```

```python
# apply steming function
final_df['tokenized'] = final_df['Text'].apply(stemming)
```
